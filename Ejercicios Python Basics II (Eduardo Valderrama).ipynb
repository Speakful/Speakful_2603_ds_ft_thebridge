## Ejercicio 2
Escribe un programa para pasar de grados a radianes. Hay que usar `input`. Recuerda que la conversión se realiza mediante

radianes = grados*(pi/180)

````python
print("Este programa convierte grados a radianes")
print("Introduzca los grados a convertir")
grados = float(input())

pi = 3.14159265359

radianes = grados*(pi/180)

print(f"{grados} grados = {radianes} radianes")
````

## Ejercicio 6
Mismo programa que en 5, pero en esta ocasión tienen que ser tres inputs y dos salidas. Una de las salidas que nos indique si todos son iguales, y la otra si al menos dos inputs sí que lo son

````python
a = input("Nombre: ")
b = input("Primer apellido: ")
c = input("Segundo apellido: ")

if a == b == c:
  print("Todos son iguales")
else:
  print("No todos son iguales")

if a == b or a == c or b == c:
  print("Al menos 2 son iguales")
else:
  print("Al menos 2 NO son iguales")
````

## Ejercicio 7
Crea un programa que recoja dos inputs. Tiene que comprobar si su suma es igual, superior o inferior a 10

````python
primer_input = int(input("Edad: "))
segundo_input = int(input("Año nacimiento: "))

suma = primer_input + segundo_input

if suma == 10:
    print(suma, "es igual a 10")

elif suma < 10:
    print(suma, "es menor que 10")

else:
    print(suma, "es mayor que 10")


````

## Ejercicio 10
Para el siguiente string se pide imprimir por pantalla los siguientes casos

    "En un lugar de la Mancha, de cuyo nombre no quiero acordarme."

1. Pásalo todo a mayúsculas
2. Pásalo todo a minúsculas
3. Solo la primera letra de cada palabra en mayúscula, el resto en minúscula
4. Crea una lista dividiéndolo por sus espacios
5. Sustituye las comas `,` por puntos y comas `;`
6. Elimina las `a` minúsculas

````python
print("En un lugar de la Mancha, de cuyo nombre no quiero acordarme.".upper())

print("En un lugar de la Mancha, de cuyo nombre no quiero acordarme.". lower())

print(("En un lugar de la Mancha, de cuyo nombre no quiero acordarme.").title())

print("En un lugar de la Mancha, de cuyo nombre no quiero acordarme.".split())

print("En un lugar de la Mancha, de cuyo nombre no quiero acordarme.".replace("," , "."))

print("En un lugar de la Mancha, de cuyo nombre no quiero acordarme.".replace("a" , ""))
````
