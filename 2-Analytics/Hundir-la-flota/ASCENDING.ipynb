```python
# Step 1: CREATE A BOARD

import numpy as np
import random

tablero = np.full((10, 10)," ")
print(tablero)
```

```python
def crea_tablero(lado = 10):
    tablero = np.full((lado, lado), " ")
    return tablero
```

```python
tablero = crea_tablero(10)
print(tablero)
```

```python
# Step 2: VALIDATE AND PLACE SHIP

def coloca_barco_plus(tablero, barco): #COLOCA Y VALIDA SI SE PUEDE UBICAR EL BARCO EN (X,Y)

    tablero_temp = tablero.copy()
    num_max_filas = tablero.shape[0]
    num_max_columnas = tablero.shape[1]

    for pieza in barco:
        fila = pieza[0]
        columna = pieza[1]

        if fila < 0  or fila >= num_max_filas:
            print(f"Coordenadas no válidas {pieza}, hay otro barco")
            return False
        
        if columna <0 or columna>= num_max_columnas:
            print(f"Coordenadas no válidas {pieza}, hay otro barco")
            return False
        
        if tablero[pieza] == "O" or tablero[pieza] == "X":
            print(f"Coordenadas no válidas {pieza}, hay otro barco")
            return False
        
        tablero_temp[pieza] = "O"
    return tablero_temp
```

```python
# Step 3: GENERATE A RANDOM SHIP

def crea_barco_aleatorio(tablero,eslora = 4, num_intentos = 100): #CREA EL BARCO
    num_max_filas = tablero.shape[0]
    num_max_columnas = tablero.shape[1]

    while True:
       
        barco = []

        pieza_original = (random.randint(0,num_max_filas-1),random.randint(0, num_max_columnas -1))
        print("Navegando hacia las coordenadas:", pieza_original)
        barco.append(pieza_original)
        
        orientacion = random.choice(["N","S","O","E"])
        if orientacion == "N":
            print("⬆️ Proa al norte, a la conquista!")
        elif orientacion == "S":
            print("⬇️ Rumbo sur, hacia aguas desconocidas!")
        elif orientacion == "O":
            print("⬅️ Virando al oeste, tierra a la vista!")
        else: 
            print("➡️ Navegando hacia el este, al horizonte!")
            
        fila = pieza_original[0]
        columna = pieza_original[1]

        for i in range(eslora -1):

            if orientacion == "N":
                fila -= 1

            elif orientacion  == "S":
                fila += 1

            elif orientacion == "E":
                columna += 1

            else:
                columna -= 1

            pieza = (fila,columna)
            barco.append(pieza)

        tablero_temp = coloca_barco_plus(tablero, barco) #PRIMERO CREAMOS LA FUNCIÓN DE COLOCAR BARCO PORQUE LA FUNCIÓN DE CREAR BARCO LLAMA A ESTA PRIMERA PARA VALIDAR SI LAS COORDENADAS SON VÁLIDAS O NO
        
        if type(tablero_temp) == np.ndarray:
            return tablero_temp
        print("Intenta colocar otro barco")

        
```

```python
# Step 4: HANDLE PLAYER SHOTS

def recibir_disparo(tablero, coordenada):
    if tablero[coordenada] != " " and tablero[coordenada] != "X" and tablero[coordenada] != "-":
        tablero[coordenada] = "X"
        print("Impacto directo, capitán! :)")
        return tablero
    elif tablero[coordenada] == "X" or tablero[coordenada] == "-":
        print("Ya disparaste aquí... Perdiste el mapa ?")
        return None
    else:
        tablero[coordenada] = "-"
        print("Agua... El mar se traga tu bala sin piedad :(")
        return tablero
```
