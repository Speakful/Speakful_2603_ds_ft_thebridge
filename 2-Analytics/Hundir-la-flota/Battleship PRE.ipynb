```python
# CREAMOS UN TABLERO 10x10 RELLENO DEL CARÁCTER ESPACIO " "

import numpy as np
import random

tablero = np.full((10,10)," ")
print(tablero)
```

```python
def crea_tablero(lado = 10):
    tablero = np.full((lado,lado)," ")
    return tablero
```

```python
tablero = crea_tablero(10)
print(tablero)
```

```python
# POSICIONA UN PAR DE BARCOS EN [(0,1),(1,1)] & [(1,3), (1,4), (1,5), (1,6)] LOS BARCOS SERÁN Os MAYÚSCULAS. COMO VES, UN BARCO DE 2 POSICIONES DE ESLORA Y OTRO DE 4


barco1 = [(0,1),(1,1)]
barco2 = [(1,3), (1,4), (1,5), (1,6)]

def coloca_barco(tablero, barco):
    for pieza in barco:
        tablero[pieza] = "O"
    return tablero


tablero = coloca_barco(tablero, barco1)
tablero = coloca_barco(tablero, barco2)
print(tablero)
```

```python
# CREAMOS EL TABLERO DE NUEVO (DESDE CERO) PARA VER EL PASO A PASO 

tablero = crea_tablero(10)
barco1 = [(0,1),(1,1)]
barco2 = [(1,3), (1,4), (1,5), (1,6)]
print("")
print(tablero)
print("")
for barco in [barco1, barco2]:
    print("")
    print("="*42)
    print("")
    print("Coloco:", barco)
    print("")
    tablero = coloca_barco(tablero,barco)
    print(tablero)
```

```python
# RECIBE UN DISAPRO EN UNO DE LOS BARCOS, SUSTITUYENDO LA 0 POR UNA X

def recibir_disparo(tablero, coordenada):
    if tablero[coordenada] == "O":
        tablero[coordenada] = "X"
        print("Impacto !")
    elif tablero[coordenada] == "X":
        print("Ya has disparado aquí... Espabila")
    else:
        tablero[coordenada] = "-"
        print("Agua :( ! ")

recibir_disparo(tablero,(5,6))
print("")
print(tablero)
```

```python
# FÍJATE EN EL GUION SI SE DISPARA SOBRE AGUA

recibir_disparo(tablero,(1,1))
print("")
print(tablero)
```

```python
# FUNCIÓN PARA ARREGLAR EL BARCO DONDE LA CELDA == X Y MODIFICARLA A "O"
def arregla_barcos(tablero):
    tablero[tablero == "X"] = "O"
    tablero[tablero == "-"] = " "
    #tablero.where(tablero == "X","O", tablero)
    return tablero

print(arregla_barcos(tablero))
```

```python
# NOS DEVUELVE SI EL TABLERO PUEDE COLOCAR EL BARCO // SI NO SE PUEDE -> DEVUELVE FALSE Y AVISA AL USUARIO POR PANTALLA

def coloca_barco_plus(tablero, barco):
    # Nos devuelve el tablero si puede colocar el barco, si no devuelve False, y avise por pantalla
    tablero_temp = tablero.copy()
    num_max_filas = tablero.shape[0]
    num_max_columnas = tablero.shape[1]
    for pieza in barco:
        fila = pieza[0]
        columna = pieza[1]
        if fila < 0  or fila >= num_max_filas:
            print(f"Las coordenadas {pieza} no son válidas")
            return False
        if columna <0 or columna>= num_max_columnas:
            print(f"Las coordenadas {pieza} no son válidas")
            return False
        if tablero[pieza] == "O" or tablero[pieza] == "X":
            print(f"Las coordenadas {pieza} no son válidas")
            return False
        tablero_temp[pieza] = "O"
    return tablero_temp
```

```python
def crea_barco_aleatorio(tablero,eslora = 4, num_intentos = 100):
    num_max_filas = tablero.shape[0]
    num_max_columnas = tablero.shape[1]
    while True:
        barco = []

        # CONSTRUIMOS NUESTRO BARCO

        pieza_original = (random.randint(0,num_max_filas-1),random.randint(0, num_max_columnas -1))
        print("Pieza original:", pieza_original)
        barco.append(pieza_original)
        orientacion = random.choice(["N","S","O","E"])
        print("Con orientacion", orientacion)
        fila = pieza_original[0]
        columna = pieza_original[1]
        for i in range(eslora -1):
            if orientacion == "N":
                fila -= 1
            elif orientacion  == "S":
                fila += 1
            elif orientacion == "E":
                columna += 1
            else:
                columna -= 1
            pieza = (fila,columna)
            barco.append(pieza)
        tablero_temp = coloca_barco_plus(tablero, barco)
        if type(tablero_temp) == np.ndarray:
            return tablero_temp
        print("Intenta colocar otro barco")
```
